```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Iterate through the list of tuples (each tuples contain two elements: lats and Lngs)
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Base URL
url= "http://api.openweathermap.org/data/2.5/weather?"
#imperial units is Fahrenheit
units="imperial"

#Build query URL (Use f string to do string concatenation)
query_url=f"{url}appid={api_key}&units={units}&q="
query_url
```

```python
#Print out the JSON object for New York for now to see the structure of the response
response=requests.get(query_url+"New York").json()
pprint(response)
```

```python
# Create several empty list to begin with
city_list=[]
city_country=[]
city_temperature=[]
city_humidity=[]
city_cloudiness=[]
city_windspeed=[]
city_latitude=[]
```

```python
count=1
for city in cities:
    print(f"Retrieving result of city {city}.")
    response=requests.get(query_url+city).json()
    response
    try:
        city_list.append(response["name"])
        city_country.append(response["sys"]["country"])
        city_temperature.append(response["main"]["temp"])
        city_humidity.append(response["main"]["humidity"])
        city_cloudiness.append(response["clouds"]["all"])
        city_windspeed.append(response["wind"]["speed"])
        city_latitude.append(response["coord"]["lat"])
        print(f"Processing data for index {count}.")
        count+=1
    
    except KeyError:
        print("Missing result...skipping.")
    
    
    print("-----------------------------------------------")
print("Retrival Complete.")
print("-----------------------------------------------")
```

```python
len(city_latitude)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Conver the multiple lists above into a data frame
Weather_df=pd.DataFrame({"City":city_list,"Country":city_country,"Temperature":city_temperature,
                         "Humidity":city_humidity,"Cloudiness":city_cloudiness,
                         "Wind Speed(mph)":city_windspeed,"Latitude":city_latitude})
Weather_df.head()
```

```python
Weather_df.count()
```

```python
#Save the dataframe to csv file
Weather_df.to_csv("Weather dataframe.csv")
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.

#### Latitude vs. Temperature Plot

```python
Latitude=Weather_df["Latitude"].tolist()
Temperature=Weather_df["Temperature"].tolist()
plt.scatter(Latitude,Temperature,marker="o",facecolors="red",edgecolors="black",alpha=0.75)
plt.grid()
plt.title("Latitude vs. Temperature")
plt.xlabel("Latitude")
plt.ylabel("Temperature(Fehrenheit)")
plt.savefig("Image/Latitude_vs_Temperature")
```

#### Latitude vs. Humidity Plot

```python
Latitude=Weather_df["Latitude"].tolist()
Humidity=Weather_df["Humidity"].tolist()

plt.scatter(Latitude,Humidity,marker="o",facecolors="blue",edgecolors="black",alpha=0.75)
plt.grid()
plt.title("Latitude vs. Humidity Percentage")
plt.xlabel("Latitude")
plt.ylabel("Humidity%")
plt.savefig("Image/Latitude_vs_Humidity")
```

#### Latitude vs. Cloudiness Plot

```python
Latitude=Weather_df["Latitude"].tolist()
Cloudiness=Weather_df["Cloudiness"].tolist()
plt.scatter(Latitude,Cloudiness,marker="o",facecolors="yellow",edgecolors="black",alpha=0.75)
plt.grid()
plt.title("Latitude vs. Cloudiness Percentage")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness%")
plt.savefig("Image/Latitude_vs_Cloudiness")
```

#### Latitude vs. Wind Speed Plot

```python
Latitude=Weather_df["Latitude"].tolist()
Wind_speed=Weather_df["Wind Speed(mph)"].tolist()
plt.scatter(Latitude,Wind_speed,marker="o",facecolors="orange",edgecolors="black",alpha=0.75)
plt.grid()
plt.title("Latitude vs. Wind Speed (mph)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.savefig("Image/Latitude_vs_WindSpeed")
```
